```
# Here we're considering staging layer data as our source and bikes_history table as our target
# Step 1 left join 
# Considering ID column as our primary key and to avoid complexity the demo selecting 'id','description','frame_model','manufacturer_name','serial' these columns

from datetime import datetime
from pyspark.sql.functions import lit ,col , current_timestamp

source_table = spark.read.table("bike_details_staging").filter("partiton_date == '{}'".format(datetime.today().strftime('%Y-%m-%d'))).select('id','description','frame_model','manufacturer_name','serial')
target_table = spark.read.table("bikes_history").select('id','description','frame_model','manufacturer_name','serial',col("partiton_date").alias("start_date"),lit("Y").alias('active_status'),lit("9999-12-31").cast("date").alias("end_date"))

base_path = "/FileStore/tables/"
target_table_path = base_path + "full_history/"
target_table.write.format("delta").mode("append").save(target_table_path)
from delta import DeltaTable

# Create Delta From Target Table as we want to use this in Merge statement

target_table = DeltaTable.forPath(spark,target_table_path)
target_df = target_table.toDF().select('id','description','frame_model','manufacturer_name','serial', current_timestamp().cast("date").alias("start_date"),lit("Y").alias('active_status'),lit("9999-12-31").cast("date").alias("end_date"))
```

```
# step 1 left join source and target

dataframe_join = source_table.join(target_df, on=("id"),how="left_outer").select(source_table["*"],target_df.id.alias("target_id"),target_df.description.alias("target_description"),target_df.frame_model.alias("target_frame_model"),target_df.manufacturer_name.alias("target_manufacturer_name"),target_df.serial.alias("target_serial"))
```

```
dataframe_join.show(10,False)
```

```
from pyspark.sql.functions import sha2, concat_ws
# Filter out only changed Records and NEW records
dataframe_filter = dataframe_join.filter(sha2(concat_ws("||", dataframe_join.id,dataframe_join.description,dataframe_join.frame_model,dataframe_join.manufacturer_name,dataframe_join.serial), 256) != 
                                        sha2(concat_ws("||", dataframe_join.target_id,dataframe_join.target_description,dataframe_join.target_frame_model,dataframe_join.target_manufacturer_name,dataframe_join.target_serial), 256)
                                        )
```

```
# Create Key column

dataframe_filter = dataframe_filter.withColumn("KEY",col("id"))



# take out changed records filter dataframe

dataframe_filter_new_records = dataframe_filter.filter("target_id is not null").withColumn("KEY",lit(None))
```

```
dataframe_filter_new_records.show(10,False)
```

```
# Combine Above Two Dataframes

dataframe_combined = dataframe_filter.union(dataframe_filter_new_records)
```

```
# Execute the merge into target table and dataframe (containing new and changed records)

target_table.alias("target").merge(source=dataframe_combined.alias("source"),condition="target.id = source.KEY and target.active_status = 'Y'"\
                                  ).whenMatchedUpdate(set = {"active_status" : "'N'","end_date" : "current_date"} 
                                  ).whenNotMatchedInsert(values = {"id":"source.id","description":"source.description","frame_model":"source.frame_model","manufacturer_name":"source.manufacturer_name","serial":"source.serial","active_status":"'Y'","start_date":"current_date","end_date":"""to_date('9999-12-31','yyyy-MM-dd')"""}).execute()
            
```
