```

# Define All the Constants

base_path = "/FileStore/tables/"
raw_layer_path = base_path + "raw_layer/"
api_endpoint = "https://bikeindex.org/api/v3/search"
```

```
# Read Data From and Store this in json File
# Read data from json file and store this in dataframe

import requests
response = requests.request('GET', api_endpoint)
dbutils.fs.put(base_path+'api_response.json', response.text)
dataframe_raw = spark.read.option('multiline', 'true').json(base_path+'api_response.json')
```

```
# As the dataframe has only one column "bikes" ,
# Added current_timestamp and  partition date column 

from pyspark.sql.functions import lit ,current_timestamp
dataframe_raw = dataframe_raw.select(
        lit(api_endpoint).alias("datasource"), 
        current_timestamp().alias("create_timestamp"), 
        "bikes",
        current_timestamp().cast("date").alias("partiton_date"))


dataframe_raw.show(10,False)
```

```
# Write data into source location in parquet format in append mode

from delta.tables import DeltaTable
from pyspark.sql import DataFrame, DataFrameWriter


dataframe_raw.write.format("delta").mode("append").partitionBy("partiton_date").save(raw_layer_path)
```

```
dbutils.fs.rm(base_path+'api_response.json', recurse=True)
```

```
# adding data to Table 
spark.sql(
    """
      DROP TABLE IF EXISTS {}
    """.format('bike_details_raw')
)

spark.sql(
  """
    CREATE TABLE {}
    USING DELTA 
    LOCATION "{}"
  """.format('bike_details_raw', raw_layer_path)
)
```
